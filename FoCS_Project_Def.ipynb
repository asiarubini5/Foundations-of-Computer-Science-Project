## 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

- Defining a list of country codes and using a for-loop to iterate through them. 
- For each country, reading the corresponding CSV file and immediately tagging it by adding a new 'country' column. 
- Finally, collecting all the dataframes and concatenating them into a single dataset

```python
country = ['CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU', 'US']
list_total = []
for c in country:
    path = '/Users/asiarubini/Desktop/UNI/DATA SCIENCE/COMPUTER SCIENCE/PROJECT FOCS/trendingYT/' + c + 'videos.csv'
    df = pd.read_csv(path, encoding_errors='replace')
    df['country'] = c
    list_total.append(df)

df_total = pd.concat(list_total, ignore_index=True)
df_total
```

## 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.


Determining the number of videos with comments disabled, ratings disabled, and those flagged as video error or removed

```python
int((df_total['comments_disabled'] == True).sum())
```

```python
int((df_total['ratings_disabled'] == True).sum())
```

```python
int((df_total['video_error_or_removed'] == True).sum())
```

Creating a new dataframe called excluded filtering all the rows with disabled comments/ratings or videos with errors or removed (using the & and | operators)

```python
excluded = df_total[
    (df_total['comments_disabled'] == True) & (df_total['ratings_disabled'] == True) |
    (df_total['video_error_or_removed'] == True)
]
```

```python
excluded
```

Saving in a new dataframe all the videos which are not in excluded

```python
df_total_noexcluded = df_total[~df_total.index.isin(excluded.index)].copy()
df_total_noexcluded
```

## 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes


```python
df_total_noexcluded['like_ratio'] = df_total_noexcluded['likes']/df_total_noexcluded['dislikes']
df_total_noexcluded
```

## 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)


Checking the format of the values in the 'publish_time' column.

```python
df_total_noexcluded['publish_time']
```

Since the values are strings, it needs to transform their format into datetime64 format

```python
df_total_noexcluded['publish_time'] = pd.to_datetime(df_total_noexcluded['publish_time'])
df_total_noexcluded['publish_time']
```

- Rounding the publish_time down to 10-minute intervals
- extracting only the time component, removing the specific date

```python
# df_total_noexcluded['publish_interval_10min'] = df_total_noexcluded['publish_time'].dt.floor('10min')
# df_total_noexcluded['publish_interval_10min'] = df_total_noexcluded['publish_interval_10min'].dt.time
```

```python
start_time = df_total_noexcluded['publish_time'].dt.floor('10min')
end_time = start_time + pd.Timedelta(minutes=10)
df_total_noexcluded['publish_interval_10min'] = ('from ' + start_time.dt.strftime('%H:%M') + ' to ' + end_time.dt.strftime('%H:%M'))
```

```python
df_total_noexcluded
```

## 7. For each interval, determine the number of videos, average number of likes and of dislikes.


```python
df_intervals = df_total_noexcluded.groupby('publish_interval_10min').agg(
    n_videos=('video_id', 'count'),      
    avg_likes=('likes', 'mean'),         
    avg_dislikes=('dislikes', 'mean')    
)
df_intervals
```

## 8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.



Checking, randomly, the format of the values in 'tags'

```python
df_total_noexcluded['tags'].sample()
```

Creating a new dataframe with all the videos which have real tags (not '[none]')

```python
df_tags = df_total_noexcluded[~(df_total_noexcluded['tags'] == '[none]')].copy()
```

```python
df_tags
```

Cleaning the column 'tags': how?
- creating a new column 'tags_cleaned' with tags values without any special characters

```python
df_tags["tags_cleaned"] = df_tags["tags"].str.replace(r'[\"]', '', regex=True)
```

- Transforming each concatenated string in 'tags_cleaned' into a list of tags saved in a new column called 'tags_splitted'


```python
df_tags["tags_splitted"] = df_tags["tags_cleaned"].str.split('|')
df_tags
```

Expanding each n-list into n-rows. Each row, in the column 'tags_splitted', will have a single tag value

```python
df_tags_exploded = df_tags.explode("tags_splitted").copy()
df_tags_exploded
```

After cleaned the tags with white spaces inside, or empty tags or [none] tags, determining the number of videos for each tag

```python
df_tags_exploded = df_tags_exploded[(df_tags_exploded['tags_splitted'] != '[none]') & (df_tags_exploded['tags_splitted'] != '')]
```

```python
df_tags_exploded.groupby('tags_splitted')['video_id'].count().sort_values(ascending=False)
```

Converting the column to datetime. The format '%y.%d.%m' tells Python to parse the date structure as Year.Day.Month.

Creating the three columns

```python
df_total_noexcluded['year'] = df_total_noexcluded['trending_date_dt'].dt.year
df_total_noexcluded['month'] = df_total_noexcluded['trending_date_dt'].dt.month
df_total_noexcluded['day'] = df_total_noexcluded['trending_date_dt'].dt.day
df_total_noexcluded
```

## 14. Read all json files with the video categories


```python
country = ['CA', 'DE', 'FR', 'GB', 'IN', 'JP', 'KR', 'MX', 'RU', 'US']
list_total_cat = []
for c in country:
    path = '/Users/asiarubini/Desktop/UNI/DATA SCIENCE/COMPUTER SCIENCE/PROJECT FOCS/trendingYT/' + c + '_category_id.json'
    df = pd.read_json(path, encoding_errors='replace')
    df_norm = pd.json_normalize(df['items'])
    df_norm['country'] = c
    df_norm['id'] = df_norm['id'].astype(int)
    list_total_cat.append(df_norm)

df_total_category = pd.concat(list_total_cat, ignore_index=True)
df_total_category
```
